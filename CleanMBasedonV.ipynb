```python
import pandas as pd
import numpy as np
df = pd.read_csv('/Users/merryxmm7/Documents/GitHub/SQL-Quest/MbasedonV.csv')
df
```

```python
# Remove everything after and including the first "[" in the "Worldwide box office" column
df['Worldwide box office'] = df['Worldwide box office'].str.split('[', n=1).str[0]
# Remove the format after the ratings in the "Rotten Tomatoes" column
df['Rotten Tomatoes'] = df['Rotten Tomatoes'].str.split('[', n=1).str[0]
# Remove the format after the ratings in the "Metacritic" column
df['Metacritic'] = df['Metacritic'].str.split('[', n=1).str[0]
```

```python
# Define the exchange rate
exchange_rate = 154.62

# Function to convert yen to dollars
def convert_to_usd(amount):
    if '¥' in str(amount):  # Convert to string to handle floats
        yen_amount = float(str(amount).replace('¥', '').replace(',', ''))
        return "${:,.2f}".format(yen_amount / exchange_rate)
    else:
        return amount

# Apply the conversion function to the "Worldwide box office" column
df['Worldwide box office'] = df['Worldwide box office'].apply(convert_to_usd)
```

```python


# Fill NaN values with "UNKNOWN" for specified columns
df['Title'].fillna('UNKNOWN', inplace=True)
df['Direction'].fillna('UNKNOWN', inplace=True)
df['Distributor'].fillna('UNKNOWN', inplace=True)
df['Original game publisher'].fillna('UNKNOWN', inplace=True)

# Fill NaN values with 0 for "Rotten Tomatoes" and "Metacritic" columns
df['Rotten Tomatoes'].fillna('0%', inplace=True)
df['Metacritic'].fillna('0/100', inplace=True)

# Remove percentage sign from "Rotten Tomatoes" column and convert to numeric type
df['Rotten Tomatoes'] = df['Rotten Tomatoes'].str.rstrip('%').replace('N/A', np.nan).astype(float)

# Remove everything after "/" from "Metacritic" column and convert to numeric type
def extract_metacritic(score):
    try:
        return int(score.split('/')[0])
    except (AttributeError, ValueError):
        return np.nan

df['Metacritic'] = df['Metacritic'].apply(extract_metacritic)
```
