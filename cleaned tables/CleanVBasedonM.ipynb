```python
import pandas as pd
import numpy as np
import unicodedata
df_vbasedm_dirty = pd.read_csv('/Users/merryxmm7/Documents/GitHub/SQL_Quest_VideoGames/mg_cleanedSL.csv')
df_vbasedm_dirty
```

```python
# Lowercase all column names
df_vbasedm_dirty = df_vbasedm_dirty.rename(columns={col: col.lower() for col in df_vbasedm_dirty.columns})
df_vbasedm_dirty
```

```python
nan_values = df_vbasedm_dirty.isna()

# Summing up the True values (NaNs) in each column
nan_count_per_column = nan_values.sum()

print("Number of NaN values per column:")
print(nan_count_per_column)
```

```python
# Fill NaN values in 'developer' and 'publisher' columns with "UNKNOWN"
df_vbasedm_dirty['developer'].fillna("UNKNOWN", inplace=True)
df_vbasedm_dirty['publisher'].fillna("UNKNOWN", inplace=True)
```

```python
# Function to check if a string contains non-ASCII characters
def contains_non_ascii(s):
    return any(ord(char) > 127 for char in s)

# Check for non-ASCII values in the DataFrame
non_ascii_values = df_vbasedm_dirty.applymap(lambda x: contains_non_ascii(str(x)))

# Find the rows and columns where non-ASCII values are present
rows_with_non_ascii = non_ascii_values.any(axis=1)
cols_with_non_ascii = non_ascii_values.any(axis=0)

# Print the rows and columns where non-ASCII values are present
print("Rows with non-ASCII values:")
print(df_vbasedm_dirty[rows_with_non_ascii])

print("\nColumns with non-ASCII values:")
print(df_vbasedm_dirty.columns[cols_with_non_ascii])
```

```python
import unicodedata

# Function to replace non-ASCII characters with their ASCII equivalents
def replace_non_ascii(s):
    try:
        return ''.join(char if unicodedata.name(char).startswith('LATIN') else '' for char in str(s))
    except ValueError:
        return ''  # Return an empty string if no Unicode name is found for the character

# Iterate over each cell containing non-ASCII characters
for col in df_vbasedm_dirty.columns:
    df_vbasedm_dirty[col] = df_vbasedm_dirty[col].apply(replace_non_ascii)

# Print the DataFrame after conversion
print("DataFrame with non-ASCII characters replaced:")
print(df_vbasedm_dirty)
```

```python
import unicodedata

# Function to replace non-ASCII characters with their ASCII equivalents
def replace_non_ascii(s):
    return ''.join(char if unicodedata.name(char).startswith('LATIN') else char.encode('ascii', 'ignore').decode('ascii') for char in str(s))

# Apply the function to the columns with non-ASCII values
for col in df_vbasedm_dirty.columns[cols_with_non_ascii]:
    df_vbasedm_dirty[col] = df_vbasedm_dirty[col].apply(replace_non_ascii)

# Print the DataFrame after conversion
print("DataFrame with non-ASCII characters replaced:")
print(df_vbasedm_dirty)
```

```python
import re

# Function to remove non-ASCII characters from a string
def remove_non_ascii(s):
    return re.sub(r'[^\x00-\x7F]+', '', str(s))

# Apply the function to the columns with non-ASCII values
for col in df_vbasedm_dirty.columns[cols_with_non_ascii]:
    df_vbasedm_dirty[col] = df_vbasedm_dirty[col].apply(remove_non_ascii)

# Print the DataFrame after conversion
print("DataFrame with non-ASCII characters replaced:")
print(df_vbasedm_dirty)
```
