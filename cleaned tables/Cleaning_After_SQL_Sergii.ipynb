```python
import pandas as pd
df= pd.read_csv(r'C:\Users\38095\Documents\GitHub\ProjectSQL\sales_on_mov.csv')
df
```

```python
df1= pd.read_csv(r'C:\Users\38095\Documents\GitHub\ProjectSQL\all_vgs_comparison.csv')
df1
```

```python
visual_df = pd.merge(df1,df, on=['year'], how = 'inner')
visual_df['average_sales_na']=visual_df['average_sales_na']*1000000
visual_df['average_sales_na'] = visual_df['average_sales_na'].astype(int)
visual_df['average_sales_jp']=visual_df['average_sales_jp']*1000000
visual_df['average_sales_jp'] = visual_df['average_sales_jp'].astype(int)
visual_df['sales_na_on_moovie']=visual_df['sales_na_on_moovie']*1000000
visual_df['sales_na_on_moovie'] = visual_df['sales_na_on_moovie'].astype(int)
visual_df['sales_jp_on_moovie']=visual_df['sales_jp_on_moovie']*1000000
visual_df['sales_jp_on_moovie'] = visual_df['sales_jp_on_moovie'].astype(int)
visual_df['sales_global_on_moovie']=visual_df['sales_global_on_moovie']*1000000
visual_df['sales_global_on_moovie'] = visual_df['sales_global_on_moovie'].astype(int)

visual_df['share_sales_na_in_global']=visual_df['share_sales_na_in_global']*100
visual_df['share_sales_na_in_global'] = visual_df['share_sales_na_in_global'].astype(int)
visual_df['share_sales_jp_in_global']=visual_df['share_sales_jp_in_global']*100
visual_df['share_sales_jp_in_global'] = visual_df['share_sales_jp_in_global'].astype(int)

visual_df['share_sales_na_on_moovie_in_global']=visual_df['share_sales_na_on_moovie_in_global']*100
visual_df['share_sales_na_on_moovie_in_global'] = visual_df['share_sales_na_on_moovie_in_global'].astype(int)
visual_df['share_sales_jp_on_moovie_in_global']=visual_df['share_sales_jp_on_moovie_in_global']*100
visual_df['share_sales_jp_on_moovie_in_global'] = visual_df['share_sales_jp_on_moovie_in_global'].astype(int)

visual_df.rename(columns={'average_sales_na': 'Noth_America_sales', 'average_sales_jp': 'Japan_sales', 'sales_na_on_moovie': 'Noth_America_sales_moovie_based', 'sales_jp_on_moovie': 'Japan_sales_moovie_based','sales_global_on_moovie': 'Global_sales_moovie_based' }, inplace=True)

visual_df
```

```python
visual_df.to_csv('visual_sldf.csv', index=False, sep=';')
```
